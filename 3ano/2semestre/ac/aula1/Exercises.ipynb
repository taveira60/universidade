### 3. CompScience FizzBuzz
Print numbers from 1 to 30, but:  
- Replace multiples of 3 with "Comp"  
- Replace multiples of 5 with "Science"  
- Replace multiples of both with "CompScience"

```python
def ex3():
    for i in range(1,30):
        if i % 3 == 0:
            print(f"Comp {i}")
            
        if i % 5 == 0:
            print(f"Science {i}")
            
        if i % 3 == 0 and i % 5 == 0:
            print(f"Comp Science {i}")
            
ex3()
            
```

Provide two solutions:
A) One using a for loop.

- $1! = 1$

- $4! = 4 \times 3 \times 2 \times 1 = 24$
- $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$

```python
def s1(x):
    a=1
    for i in range (1,x+1):
        a=a*i
    print(a)
s1(4)

def s2(x):
    a=1
    i=1
    while i != x+1:
        a=a*i
        i+=1
    print(a)
s2(4)
```

### 1. Palindrome
Write `is_palindrome(word)` returning `True` if `word` reads the same backwards.  

```python
def is_palindrome(word):
    i = 0
    j = len(word) - 1
    while i < j:
        if word[i] != word[j]:
            return False
        i += 1
        j -= 1
    return True

print(is_palindrome("aaaabaaaa"))
```
